### 이름이 똑같은데 product_id는 다른 경우가 새상품과 중고를 나눈 것이다. 가설 검정

```python
# Grouping by productDisplayName and product_id to get the descriptive statistics for item_price
price_stats = dup_category_second.groupby(['productDisplayName', 'product_id'])['item_price'].describe()

# Filtering for productDisplayName with more than one product_id
multiple_products = price_stats.groupby(level=0).filter(lambda x: len(x) > 1)

# Correcting the error and separating prices of used and new products based on the criteria mentioned
used_prices = []
new_prices = []

for product_name, group in multiple_products.groupby(level=0):
    min_price_product_id = group['mean'].idxmin()[1]
    used_prices.append(group.loc[(product_name, min_price_product_id), 'mean'])
    
    other_product_ids = group.drop(min_price_product_id, level=1)
    new_prices.extend(other_product_ids['mean'].values)

# Convert lists to series for further analysis
used_prices_series = pd.Series(used_prices, name="Used Product Prices")
new_prices_series = pd.Series(new_prices, name="New Product Prices")

from scipy.stats import ttest_ind

# Conducting the independent two-sample t-test
t_stat, p_value = ttest_ind(new_prices_series, used_prices_series, alternative='greater')

t_stat, p_value





```

# 가설 채택, 중고-새상품 컬럼 생성

```python
dup_category_second
```

```python
# Create a mapping for product_id to product status based on the criteria
product_status_mapping = {}

# For each productDisplayName with multiple product_ids
for product_name, group in multiple_products.groupby(level=0):
    min_price_product_id = group['mean'].idxmin()[1]
    product_status_mapping[min_price_product_id] = 'second_hand'
    
    other_product_ids = group.drop(min_price_product_id, level=1).index.get_level_values('product_id').tolist()
    for pid in other_product_ids:
        product_status_mapping[pid] = 'brand_new'

# Map product_status to the original dataframe
dup_category_second['product_status'] = dup_category_second['product_id'].map(product_status_mapping).fillna('unknown')

# Display the updated dataframe
dup_category_second.head()
```
